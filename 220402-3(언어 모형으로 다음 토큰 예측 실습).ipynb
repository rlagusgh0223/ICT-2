```python
import tensorflow as tf
# AutoTokenizer : 미리 학습된 토크나이저
# TFAutoModelForCausalLM : Language Model 언어 모형
from transformers import TFAutoModelForCausalLM, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("xlnet-base-cased")
```

```python
model = TFAutoModelForCausalLM.from_pretrained("xlnet-base-cased")
```

```python
tokenizer.vocab_size
```

```python
vocab = tokenizer.get_vocab()
vocab
```

```python
id2word = {i:word for word, i in vocab.items()}
id2word
```

```python
sequence = f'Once upon a time, there was'
```

```python
input_ids = tokenizer.encode(sequence, return_tensors='tf')
```

```python
input_ids
```

```python
id2word[1977]
```

```python
result = model(input_ids)
```

```python
result
```

```python
logits = result[0]
logits
```

```python
logits.shape
```

```python
next_token_logits = logits[:, -1, :]
```

```python
next_token_logits
```

```python
next_token_logits.shape
```

```python
tf.nn.softmax(next_token_logits)
```

```python
top = tf.math.top_k(next_token_logits, k=10)
top
```

```python
top.indices
```

```python
for i in top.indices[0].numpy().tolist():
    print(id2word[i])
```
